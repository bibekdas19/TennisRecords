```python
HardCourt = pd.merge(MostWinnersUSOpen,MostWinnersAustralianOpen,right_index=True,left_index=True)
HardCourt
```

```python
GrassClay = pd.merge(MostWinnersWimbledon,MostWinnersFrenchOpen,right_index=True,left_index=True)
GrassClay
```

The career achievement of all four major championships in one discipline is termed a Career Grand Slam in that format. 

```python
CareerGrandSlam = pd.merge(HardCourt,GrassClay,right_index=True,left_index=True)
CareerGrandSlam=CareerGrandSlam.rename(columns={'WINNER_x_x':'US Open','WINNER_y_x':'Australian Open',
                                'WINNER_x_y':'Wimbledon','WINNER_y_y':'French Open'},inplace=False)
CareerGrandSlam['Total']=CareerGrandSlam['US Open'] + CareerGrandSlam['Australian Open'] + CareerGrandSlam['Wimbledon'] + CareerGrandSlam['French Open']
print(CareerGrandSlam)
```

So far only 5 men has been acheived the Career Grand Slam feat. It is evident that even Pete Sampras and Bjorn Berg has won grand slam 14 and 11 times, they could not achieve the Career Grand Slam feat whereas Rod Laver with only 5 total wins in open era had achieved it.
